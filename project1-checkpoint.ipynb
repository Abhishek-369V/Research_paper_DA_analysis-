 # Sentiment Analysis of 'Abstract' column:

* Applying sentiments :-

```python
# Function to analyze sentiment using TextBlob
def set_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to each abstract
df['Sentiments'] = df['Abstract'].apply(set_sentiment)
```

# Testing and Training model

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
import pickle                   #save and load machine learning models
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk
import re
# Download stopwords
nltk.download('stopwords')
```

```python
#splitting x and y 
x = df['Abstract']
y = df['Sentiments']
```

* Data Processing and applying Classification Algorithm :-

```python
def preprocess_text(text):
    # Lowercase text
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    # Stemming
    stemmer = PorterStemmer()
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text

# Apply preprocessing to the text data
x = x.apply(preprocess_text)
```

```python
# Vectorize text features using TfidfVectorizer
vectorizer = TfidfVectorizer()
x_vec = vectorizer.fit_transform(x)
```

```python
# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_vec, y, test_size=0.3, random_state=42)
```

```python
# Initialize RandomForestClassifier
model = RandomForestClassifier(random_state=42)
```

```python
# Fit model on training data
model.fit(x_train, y_train)

# Predict on testing data
y_pred = model.predict(x_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Classification report
print(classification_report(y_test, y_pred))

# Confusion matrix
print(confusion_matrix(y_test, y_pred))
```

~ Accuracy is low, due to the sentiment ratio imbalance i.e., positive sentiment is 6 times bigger than negative and 10 times more than neutral. 

```python
# Save model to file using pickle
with open('sentiment_randomforest_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Save vectorizer to file using pickle
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
```

```python
# Load model and vectorizer from file
with open('sentiment_randomforest_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)

with open('tfidf_vectorizer.pkl', 'rb') as f:
    loaded_vectorizer = pickle.load(f)

# Example prediction with loaded model and vectorizer
sample_text = "This is a best abstract text."
sample_text_preprocessed = preprocess_text(sample_text)
sample_text_vec = loaded_vectorizer.transform([sample_text_preprocessed])
sample_prediction = loaded_model.predict(sample_text_vec)
print("Sample prediction:", sample_prediction)
```

# Multiprediction Analysis:


```python
x = df['Abstract']
y_originality = df['Originality']
y_authorized = df['is_authorized']
y_trending = df['is_trending']
y_sentiment = df['Sentiments']
```

```python
def preprocess_text(text):
    # Lowercase text
    text = text.lower()
    # Remove punctuation and numbers
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    # Stemming
    stemmer = PorterStemmer()
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text

# Apply preprocessing to the text data
x = x.apply(preprocess_text)
```

```python
# Vectorize text features using TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000)
x_vec = vectorizer.fit_transform(x)
```

```python
def train_model(x, y):
    # Split data into training and testing sets using stratification
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=42)
    
```

```python
# Initialize GradientBoostingClassifier
model = GradientBoostingClassifier(random_state=42)
    
```

```python
# Fit model on training data
model.fit(x_train, y_train)
    
# Predict on testing data
y_pred = model.predict(x_test)
```

```python
# Train and evaluate model for each target
model_originality = train_model(x_vec, y_originality)
model_authorized = train_model(x_vec, y_authorized)
model_trending = train_model(x_vec, y_trending)
model_sentiment = train_model(x_vec, y_sentiment)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
    
# Classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Save models and vectorizer to files using pickle
with open('model_originality.pkl', 'wb') as f:
    pickle.dump(model_originality, f)

with open('model_authorized.pkl', 'wb') as f:
    pickle.dump(model_authorized, f)

with open('model_trending.pkl', 'wb') as f:
    pickle.dump(model_trending, f)

with open('model_sentiment.pkl', 'wb') as f:
    pickle.dump(model_sentiment, f)

with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
```
